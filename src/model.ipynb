Data cleaning

```python
# Drop null rows
df_cleaned = df.dropna()
df_cleaned.shape
```

```python
# Data type conversions
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], format='%m/%d/%Y')
# Create a new column to store yearmonth as integer
df_cleaned['yearmonth'] = df_cleaned['Date'].dt.strftime('%Y%m').astype(int)

df_cleaned
```

Linear Regression Model

```python
import statsmodels.api as sm
X = df_cleaned[['Age','Ratings']]
y = df_cleaned['Total_Amount']

# Add constant to X
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```
